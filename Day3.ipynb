```python
def topla(a,b):
  return a+b
def cikar(a,b):
  return a-b
def carp(a,b):
  return a*b
def bol(a,b):
  return a/b
def ussu_al(a,b):
  return a**b

s1 = int(input("Birinci sayıyı giriniz: "))
if s1%2 == 0:
  print("Birinci sayı çifttir.")
else:
  print("Birinci sayı tektir.")
s2 = int(input("İkinci sayıyı giriniz: "))
if s2%2 == 0:
  print("İkinci sayı çifttir.")
else:
  print("İkinci sayı tektir.")
islem = input("Yapılacak işlemi giriniz: ")

if "çarp" in islem:
  print(carp(s1,s2))
elif "topla" in islem:
  print(topla(s1,s2))
elif "çıkar" in islem:
  print(cikar(s1,s2))
elif "böl" in islem:
  print(bol(s1,s2))
elif "üssü al" in islem:
  print(ussu_al(s1,s2))
else:
  print("Lütfen tekrar deneyiniz.")
```

```python
#Hesap makinesi

def topla(a,b):
  return a+b

def cikar(a,b):
  return a-b

def bol(a,b):
  return a/b

def carp(a,b):
  return a*b

def ussu_al(a,b):
  return a**b

def kalansiz_bol(a,b):
  return a//b

def tekmiciftmi(a,b):
  if (a%2 == 0) and (b%2 == 0):
    return (a,"ve",b,"Çift Sayı")
  elif (a%2 != 0) and (b%2 != 0):
    return (a,"ve",b,"Tek Sayı")
  elif (a%2 == 0) and (b%2 != 0):
    return (a,"Çifttir",b,"Tektir")
  elif (a%2 != 0) and (b%2 == 0):
    return (a,"Tektir",b,"Çifttir")

while True:
  print("1- Toplama")
  print("2- Çıkarma")
  print("3- Bölme")
  print("4- Çarpma")
  print("5- Üssü Alma")
  print("6- Kalansiz Bölme")
  print("7- Tek mi çift mi?")
  print("8- Çıkış Yap")
  islem = int(input("Yapmak istediğiniz işlemi seçiniz: "))

  if islem == 8:
    print("Program sonlandırıldı. Güle güle...")
    break

  try:
    x = float(input("Birinci sayıyı giriniz: "))
    y = float(input("İkinci sayıyı giriniz: "))

  except:
    print("Lütfen sayıları doğru giriniz.")
    continue
  if islem == 1:
    print(topla(x,y))
  elif islem == 2:
    print(cikar(x,y))
  elif islem == 3:
    print(bol(x,y))
  elif islem == 4:
    print(carp(x,y))
  elif islem == 5:
    print(ussu_al(x,y))
  elif islem == 6:
    print(kalansiz_bol(x,y))
  elif islem == 7:
    print(tekmiciftmi(x,y))
  else:
    print("Lütfen tekrar deneyiniz.")
    continue
```

```python
# kelime sayacı

def kelimesay(yazi):
  return len(yazi.split())
def enuzunkelime(yazi):
  kelimeler = (yazi.split())
  return max(kelimeler, key=len)
def karaktersay(yazi):
  return len(yazi) - yazi.count(" ")

metin = input("Metin yazınız: ")

print("Toplam Karakter Sayısı: ", karaktersay(metin))
print("Toplam Kelime Sayısı: ", kelimesay(metin))
print("En Uzun Kelime: ", enuzunkelime(metin))
```
