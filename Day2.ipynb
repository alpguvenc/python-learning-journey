```python
sayi = int(input("Sayı girin: "))

if sayi == 156:
  print("Sayı 156'ya eşittir.")
elif sayi < 156:
  print("Sayı 156'dan küçüktür.")
else:
  print("Sayı 156'dan büyüktür.")

yas = int(input("Lütfen yaşınızı girin: "))

if yas >= 18:
  print("Reşitsiniz.")
else:
  print("Reşit değilsiniz.")
```

```python
meyveler = ["elma", "armut", "karpuz", "çilek"]

print(meyveler)
print(len(meyveler))
print(meyveler[0])
print(meyveler[1])
print(meyveler[2])
print(meyveler[3])

x = input("Meyve ekleyin: ")

meyveler.append(x)
print(meyveler)
print(len(meyveler))

y = input("Meyve silin: ")

meyveler.remove(y)
print(meyveler)
print(len(meyveler))

meyveler[0] = "muz"
print(meyveler)
print(len(meyveler))

meyveler.sort()
print(meyveler)
print(len(meyveler))

meyveler.reverse()
print(meyveler)
print(len(meyveler))
```

```python
# Not hesaplama
while True:

  not1 = int(input("Vize notunuzu girin: "))
  not2 = int(input("Final notunuzu girin: "))
  not3 = int(input("Ödev notunuzu girin: "))

  ort = not1*0.3 + not2*0.5 + not3*0.2

  print("Ortalama puanınız: ", ort)
  if ort>= 90:
    print("Harf notunuz: AA")
  elif ort>= 85:
    print("Harf notunuz: BA")
  elif ort>= 80:
    print("Harf notunuz: BB")
  elif ort>= 75:
    print("Harf notunuz: CB")
  elif ort>= 70:
    print("Harf notunuz: CC")
  elif ort>= 65:
    print("Harf notunuz: DC")
  elif ort>= 60:
    print("Harf notunuz: DD")
  elif ort>= 55:
    print("Harf notunuz: FD")
  else:
    print("Harf notunuz: FF")
  if ort == 0:
    break
```
