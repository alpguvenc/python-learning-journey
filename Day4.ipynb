```python
# hasta kayıt sistemi
L = []

while True:
  TC = int(input("TC No Giriniz: "))
  if TC in L:
    i=L.index(TC)
    print("Sıra numaranız:", i+1)
  elif TC==0:
    print(L[0], "TC nolu hasta doktorun odasına gidiniz.")
    L.pop(0)
  elif TC==-1:
    print("Sistemden çıkış yapıldı.")
    break
  else:
    L.append(TC)
    print("Sisteme kaydınız yapıldı.")
```

```python
ogrenciler = [
    {"İsim": "Alp", "Vize": 50, "Final": 70},
    {"İsim": "Buse", "Vize": 60, "Final": 80},
    {"İsim": "Orçun", "Vize": 70, "Final": 90},
    {"İsim": "İlayda", "Vize": 80, "Final": 100}

]

print(ogrenciler[0]["İsim"], "\nVize Notu:", ogrenciler[0]["Vize"], "\nFinal Notu:", ogrenciler[0]["Final"], "\nOrtalama: ", (ogrenciler[0]["Vize"]+ogrenciler[0]["Final"])/2)
print(ogrenciler[1]["İsim"], "\nVize Notu:", ogrenciler[1]["Vize"], "\nFinal Notu:", ogrenciler[1]["Final"], "\nOrtalama: ", (ogrenciler[1]["Vize"]+ogrenciler[1]["Final"])/2)
print(ogrenciler[2]["İsim"], "\nVize Notu:", ogrenciler[2]["Vize"], "\nFinal Notu:", ogrenciler[2]["Final"], "\nOrtalama: ", (ogrenciler[2]["Vize"]+ogrenciler[2]["Final"])/2)
print(ogrenciler[3]["İsim"], "\nVize Notu:", ogrenciler[3]["Vize"], "\nFinal Notu:", ogrenciler[3]["Final"], "\nOrtalama: ", (ogrenciler[3]["Vize"]+ogrenciler[3]["Final"])/2)

for ogr in ogrenciler:
    ort = (ogr["Vize"] + ogr["Final"]) / 2
    print(f"{ogr['İsim']} - Ortalaması: {ort}")
en_basarili = max(ogrenciler, key=lambda ogr: (ogr["Vize"] + ogr["Final"]) / 2)
print(f"En Başarılı Öğrenci: {en_basarili['İsim']}")
```

```python
def kayit_ol(kullanici_adi, sifre):
    with open("users.txt", "a", encoding="utf-8") as f:
        f.write(kullanici_adi + "," + sifre + "\n")
    print("Kayıt başarılı!")

def giris_yap(kullanici_adi, sifre):
    try:
        with open("users.txt", "r", encoding="utf-8") as f:
            satirlar = f.readlines()
    except FileNotFoundError:
        print("Henüz hiç kullanıcı yok. Önce kayıt olun.")
        return False

    for satir in satirlar:
        kayitli_kullanici, kayitli_sifre = satir.strip().split(",")
        if kayitli_kullanici == kullanici_adi and kayitli_sifre == sifre:
            print("Giriş başarılı. Hoş geldin", kullanici_adi)
            return True

    print("Kullanıcı adı veya şifre hatalı.")
    return False

while True:
    print("\n--- Kullanıcı Sistemi ---")
    print("1- Kayıt ol")
    print("2- Giriş yap")
    print("3- Çıkış")

    secim = input("Seçimin: ")

    if secim == "1":
        kadi = input("Kullanıcı adı: ")
        sifre = input("Şifre: ")
        kayit_ol(kadi, sifre)

    elif secim == "2":
        kadi = input("Kullanıcı adı: ")
        sifre = input("Şifre: ")
        giris_yap(kadi, sifre)

    elif secim == "3":
        print("Program sonlandırıldı.")
        break
    else:
        print("Geçersiz seçim.")
```
