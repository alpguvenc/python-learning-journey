```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df["total_bill"]

df[["total_bill","tip","sex"]].head()

df.sort_values("total_bill", ascending=False).head()

df[df["total_bill"] > 50]

df[(df["total_bill"] > 30) & (df["sex"] == "Male")]

df["total_bill"].mean()
df["tip"].max()
df["size"].value_counts()

df["tip_orani"] = df["tip"] / df["total_bill"]
df.head()

df = df.drop("size", axis=1)
df.head()
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df["tip_orani"] = df["tip"] / df["total_bill"]
df.head()

print("Ortalama Hesap:",df["total_bill"].mean())
print("En Yüksek 10 Hesap:\n",df["total_bill"].head(10))
print("Ortalama Hesap:",df.groupby("sex")["total_bill"].mean())
print("Ortalama Bahşiş:",df.groupby("smoker")["tip_orani"].mean())

print(df.groupby("day")["tip"].sum().sort_values(ascending=False).head(1))
print(df.groupby("day")["tip"].mean().sort_values(ascending=False).head(1))
```
